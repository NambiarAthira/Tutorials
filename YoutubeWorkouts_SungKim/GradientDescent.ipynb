```python
import numpy as np     # convenient matrix, tensor-like operations
import matplotlib.pyplot as plt     # plotting tool
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
w=1.0 # random wt
alpha=0.01
```

```python
#Forward pass
def forward(x):
    return x * w
```

```python
#Loss function
def loss(x, y):
    y_pred=forward(x)
    return (y-y_pred)*(y-y_pred)
```

```python
#Compute Gradient
def gradient(x, y): #d_loss/d_w
    return 2 * x * (x * w - y)
```

```python
#Before training
print("predict (before training)", 4)
```

```python
#Training loop
for epoch in range(1000):
    for x,y in zip(x_data , y_data):
        grad=gradient(x , y)
        w = w - alpha * grad
        print("\t grad:", x, y, grad)
        l=loss(x,y)
        
    print("progress:", epoch, "w=",w, "loss", l)    
```

```python
#After training
print("predict (after training)", 4)
```
