```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w_list=[]
mse_list=[]
l_sum=0
```

```python
#Model for forward pass
def forward(x):
    return x*w
```

```python
#loss  function
def loss(x,y):
    y_pred=forward(x)
    return (y-y_pred)*(y-y_pred)
```

```python
for w in np.arange(0.0, 4.1, 0.1):
    print (w)
    for x,y in zip(x_data, y_data):
        y_pred=forward(x)
        l=loss(x,y_pred)
        l_sum +=l
        print("\t",x,y,y_pred,l)
    
    print("MSE",l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3)
    
```

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
