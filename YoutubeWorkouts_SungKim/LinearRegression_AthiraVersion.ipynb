```python
# Linear regression model in PyTorch  (y=3*x)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.Tensor([[1],[2],[3]])
y = torch.Tensor([[3],[6],[9]])


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.linear=nn.Linear(1,1)   
    def forward(self,x):
        y_pred=self.linear(x)     
        return y_pred

net=Net()

# print(net)
```

```python
criterion=nn.MSELoss(size_average=False)
optimizer=optim.SGD(net.parameters(), lr=0.01)
```

```python
#Training
for epoch in range(1000):
    y_pred=net(x)
    loss=criterion(y_pred,y)
    print(epoch, loss)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
        
```

```python
#Testing
p = torch.rand(1)*100
p_pred=net(p)
print(p,p_pred)
```
